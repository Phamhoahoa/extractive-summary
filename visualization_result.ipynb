```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import test2
from math import ceil

labels = ['rouge_1', 'rouge_2', 'rouge_l']

textrank_rouge_1,textrank_rouge_2, textrank_rouge_l = test2.Test_sum().validate_sum("system/")
kmeans_rouge_1, kmeans_rouge_2, kmeans_rouge_l = test2.Test_sum().validate_sum("system_Kmeans/")
textrank = [ceil(textrank_rouge_1 * 1000)/ 1000.0,ceil(textrank_rouge_2 * 1000)/1000.0 ,ceil(textrank_rouge_l*1000)/1000.0 ]
kmeans = [ceil(kmeans_rouge_1*1000)/1000.0, ceil(kmeans_rouge_2*1000)/1000.0, ceil(kmeans_rouge_l*1000)/1000.0]
x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, textrank, width, label='textrank')
rects2 = ax.bar(x + width/2, kmeans, width, label='Kmeans')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Result Text Sumarization')
ax.set_xticks(x)
ax.set_xticklabels(labels)
plt.yticks(np.arange(0, 1.1, 0.1))
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
